```python

#linear regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
data=pd.read_csv('auto-mpg.csv')

#separating the predictor variables

predictors=data.iloc[:,1:7]

#separating the target variable

target=data.iloc[:,0]

predictors=predictors.replace("?", np.nan)

from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan, strategy="mean")

imputer.fit(predictors.iloc[:,1:6])

predictors.iloc[:,1:6]=imputer.transform(predictors.iloc[:,1:6])

predictors_train, predictors_test, target_train, target_test=train_test_split( predictors,target, test_size=0.3, random_state=121)

lr=LinearRegression()

model=lr.fit(predictors_train, target_train)

predictions=lr.predict(predictors_test)

mse=mean_squared_error(target_test, predictions)

print("mean squared error:", mse)

#Visualization

plt.scatter(target_test, predictions) #Create a scatter plot

plt.plot(target_test, target_test, color='red') #Add the regression line

plt.xlabel("Actual MPG")

plt.ylabel("Predicted MPG")

plt.title("Actual vs Predicted MPG")

plt.show()
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer

# Load the cancer dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Display the first few rows of the dataset
print(df.head())

# Splitting the dataset into training and testing sets
X = df.drop('target', axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardizing the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Applying SVM with different kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
accuracies = []
for kernel in kernels:
    # Indented block for the for loop
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"Kernel: {kernel}")
    print(f"Accuracy: {acc}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}\n")

# Visualization of the decision boundaries using PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
for kernel in kernels:
    # Indented block for the for loop
    model = SVC(kernel=kernel)
    model.fit(X_train_pca, y_train)
    # Create a mesh to plot in
    x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
    y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))
    # Plotting the decision boundary
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, edgecolors='k', marker='o')
    plt.title(f'SVM with {kernel} kernel')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()

# Summary of kernel impact on SVM performance
kernel_acc_df = pd.DataFrame({'Kernel': kernels, 'Accuracy': accuracies})
print(kernel_acc_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

def kmeans_analysis(data, dataset_name):
    # List to store inertia and silhouette scores
    inertia = []
    silhouette_scores = []
    # Testing K values from 2 to 10
    K_range = range(2, 11)
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(data)
        # Calculate inertia (within-cluster sum of squares)
        inertia.append(kmeans.inertia_)
        # Calculate silhouette score
        silhouette_avg = silhouette_score(data, cluster_labels)
        silhouette_scores.append(silhouette_avg)
    # Plot the Elbow Method
    plt.figure(figsize=(10, 5))
    plt.plot(K_range, inertia, 'bo-', label='Inertia')
    plt.title(f'Elbow Method for {dataset_name} dataset')
    plt.xlabel('Number of clusters, K')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
    # Plot Silhouette Scores
    plt.figure(figsize=(10, 5))
    plt.plot(K_range, silhouette_scores, 'ro-', label='Silhouette Score')
    plt.title(f'Silhouette Scores for {dataset_name} dataset')
    plt.xlabel('Number of clusters, K')
    plt.ylabel('Silhouette Score')
    plt.grid(True)
    plt.show()
    # Find the optimal K (maximum silhouette score)
    optimal_k = K_range[np.argmax(silhouette_scores)]
    optimal_silhouette = max(silhouette_scores)
    print(f"Optimal K for {dataset_name} dataset: {optimal_k}")
    print(f"Silhouette Score for optimal K: {optimal_silhouette}")
    return optimal_k, optimal_silhouette

# Load the Iris dataset
iris = load_iris()
iris_data = iris.data
# Load the Breast Cancer dataset
cancer = load_breast_cancer()
cancer_data = cancer.data
# Normalize the data
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris_data)
cancer_scaled = scaler.fit_transform(cancer_data)

# Run K-Means analysis on the Iris dataset
iris_k, iris_silhouette = kmeans_analysis(iris_scaled, "Iris")
# Run K-Means analysis on the Breast Cancer dataset
cancer_k, cancer_silhouette = kmeans_analysis(cancer_scaled, "Breast Cancer")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
# Creating the dataset
data = pd.read_csv('auto-mpg1.csv')


le = LabelEncoder()
data['car name'] = le.fit_transform(data['car name'])
# Impute missing values in 'horsepower' with the mean
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
data['horsepower'].fillna(data['horsepower'].mean(), inplace=True)
data['mpg_class'] = np.where(data['mpg'] < data['mpg'].median(), 'low', 'high')

# Define features (X) and target (y)
X = data.drop(columns=["origin", "mpg_class"])
y = data["origin"]

# Set seed value
seed = 123

# Splitting the dataset (70-30 and 80-20 splits)
splits = {
    "70-30": train_test_split(X, y, test_size=0.3, random_state=seed),
    "80-20": train_test_split(X, y, test_size=0.2, random_state=seed),
}

# Models: Logistic Regression and Decision Tree Classifier
models = {
    "Logistic Regression": LogisticRegression(max_iter=500, random_state=seed),
    "Decision Tree": DecisionTreeClassifier(random_state=seed),
}

# Initialize results storage
results = {}

# Training and evaluation
for split_name, (X_train, X_test, y_train, y_test) in splits.items():
    results[split_name] = {}
    for model_name, model in models.items():
        # Train the model
        model.fit(X_train, y_train)
        # Make predictions
        y_pred = model.predict(X_test)
        # Evaluate the model
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average="weighted")
        recall = recall_score(y_test, y_pred, average="weighted")
        conf_matrix = confusion_matrix(y_test, y_pred)
        # Store the results
        results[split_name][model_name] = {
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall,
            "confusion_matrix": conf_matrix,
        }

# Prepare change in accuracy table
accuracy_changes = {
    model_name: abs(
        results["70-30"][model_name]["accuracy"] - results["80-20"][model_name]["accuracy"]
    )
    for model_name in models.keys()
}

# Print results and accuracy changes
for split, metrics in results.items():
    print(f"Results for {split} split:")
    for model, values in metrics.items():
        print(f"  {model}:")
        print(f"    Accuracy: {values['accuracy']:.2f}")
        print(f"    Precision: {values['precision']:.2f}")
        print(f"    Recall: {values['recall']:.2f}")
        print(f"    Confusion Matrix:\n{values['confusion_matrix']}")
    print()

print("Change in Accuracy between Splits:")
for model, change in accuracy_changes.items():
    print(f"  {model}: {change:.2f}")
```
